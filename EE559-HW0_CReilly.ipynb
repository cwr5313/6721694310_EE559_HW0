Q7: MATPLOTLIB

```python
import matplotlib.pyplot as mp
```

```python
# Part A
x = np.arange(100)
y = np.ones(100)*np.sin(x/10)

mp.scatter(x,y)
mp.title('Scatter Plot, sin(x) v. x')
mp.xlabel('X Axis Label')
mp.ylabel('Y Axis Label')
mp.grid('on')
```

```python
# Part B
mp.plot(x,y,c='black')
mp.scatter(x,y,c='red')
mp.plot(x,np.exp(-x/10),c='blue')
mp.title('Multiple Plots, 1) sin(x) v. x, 2) exp(-x/10)')
mp.xlabel('X Axis Label')
mp.ylabel('Y Axis Label')
mp.legend(['1 Cont','1 Scat','2 Cont'],loc='lower right')
mp.grid('on')
mp.text(5,-0.65,'y1=sin(x/10)')
mp.text(5,-.9,'y2=exp(-x/10)')
```

```python
# Part C

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Plot the graphs with titles, colors in subplots
figs, axs = mp.subplots(3,1)
figs.suptitle('This is the Overarching Title, Size 20, Monospace Font',fontsize=20,family='monospace')
axs[0].plot(x,y,c='blue')
axs[0].set_title('[0,0] Title')
axs[1].plot(-x,y,c='black')
axs[1].set_title('[1,0] Title')
axs[2].plot(x,-y,c='red')
axs[2].set_title('[2,0] Title')

# Change figure sizing
figs.set_figheight(10)
figs.set_figwidth(15)


#for ax in figs.get_axes(): # moves all the axes to the outside, so no repeats.
#    ax.label_outer()
```

```python
# Polar Axes Definitions
fig, (ax1, ax2) = mp.subplots(1, 2, subplot_kw=dict(projection='polar'))
ax1.plot(x, y)
ax2.plot(x, y ** 2)

mp.show()
```

```python
# Part D

# Simple Start
mp.figure()
print(mp.gca())

# 2D Example - this example was taken from a stackoverflow forum to demonstrate 2D GCA - 'get current axes'
import matplotlib.pyplot as plt 
import numpy as np 
import matplotlib.gridspec as gridspec 
from mpl_toolkits.axes_grid1 import make_axes_locatable 
 
 
arr = np.arange(100).reshape((10, 10)) 
fig = plt.figure() 
im = plt.imshow(arr,interpolation ="none",cmap ="Blues") 
 
divider = make_axes_locatable(plt.gca()) 
cax = divider.append_axes("left", "20 %", pad ="40 %") 
 
plt.colorbar(im, cax = cax) 
 
fig.suptitle('matplotlib.pyplot.gca()', fontweight ="bold") 
plt.show() 

# 3D Example - this example was taken from a stackoverflow forum to demonstrate 3D GCA - 'get current axes'
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
 
plt.figure()
x = np.arange(-2, 2, 0.25)
y = np.arange(-2, 2, 0.25)
x, y = np.meshgrid(x, y)
z = np.sin(x * np.pi / 2) + np.cos(y * np.pi / 3)
 
fig = plt.figure()
ax = fig.gca(projection="3d")
surf = ax.plot_surface(x, y, z, cmap='Blues')
plt.show()
```
